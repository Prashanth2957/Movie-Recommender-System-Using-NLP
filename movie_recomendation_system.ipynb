```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

```

```python
movies_data=pd.read_csv('/content/tmdb_5000_movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
# selecting some relavent features for recimmendation
selected_features=['genres','keywords','tagline','cast','director','overview']
print(selected_features)
```

```python
features_to_fill = [feature for feature in selected_features if feature in movies_data.columns]

for feature in features_to_fill:
  movies_data[feature]=movies_data[feature].fillna('')
```

```python
present_features = [feature for feature in selected_features if feature in movies_data.columns]
combined_features = movies_data[present_features[0]].astype(str)
for feature in present_features[1:]:
    combined_features = combined_features + ' ' + movies_data[feature].astype(str)
```

```python
print(combined_features)
```

```python
vectorizer = TfidfVectorizer()
feature_vectors = vectorizer.fit_transform(combined_features)
print(feature_vectors)
```

```python
similarity=cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
movie_name=input('enter movie name :')
```

```python
list_of_all_titles=movies_data['title'].tolist()
print(list_of_all_titles)
```

```python
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
index_of_the_movie = movies_data[movies_data.title == close_match].index[0]
print(index_of_the_movie)
```

```python
similarity_score=list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
sorted_similarity=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similarity)
```

```python
print("Movies suggested for you :\n")

i=1

for movie in sorted_similarity:
  index = movie[0]
  title_from_index = movies_data.iloc[index]['title']
  if (i<30):
    print(i,'.',title_from_index)
    i+=1
```

```python
movie_name=input('enter movie name :')
list_of_all_titles=movies_data['title'].tolist()

find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
close_match=find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match].index[0]

similarity_score=list(enumerate(similarity[index_of_the_movie]))

sorted_similarity=sorted(similarity_score,key=lambda x:x[1],reverse=True)

print("Movies suggested for you :\n")

i=1

for movie in sorted_similarity:
  index = movie[0]
  title_from_index = movies_data.iloc[index]['title']
  if (i<30):
    print(i,'.',title_from_index)
    i+=1
```
